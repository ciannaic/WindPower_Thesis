```python
def read_and_filter_excel(file_path, specific_column, date_column):
    # Read the Excel file
    df = pd.read_csv(file_path, sep=',', skiprows = 9,  usecols=[specific_column, date_column] )

    # Filter rows based on the condition
    # filtered_df = df[df[condition_col] == condition_val]
    
    # Convert the date column to datetime format for proper sorting
    df[date_column] = pd.to_datetime(df[date_column])

    # Sort the filtered DataFrame by date
    sorted_df = df.sort_values(by=date_column)

    return sorted_df

# pattern = f"C:\\Users\\ciand\\OneDrive\\Documents\\UCD Engineering\\Final Year Project\\Penmanshiel 2\\Turbine_Data*.csv"
# # Find all files in the directory that match the pattern
# files = glob.glob(pattern)

def read_zip(file_path_zip):
    
    combined_data = pd.DataFrame()
    zf_open = zipfile.ZipFile(file_path_zip, mode = 'r')
    
    for filename in zf_open.namelist():
        if filename.startswith('Turbine'):
            
            print("filename = ", filename)
            with zf_open.open(filename) as f:
                
                 df_final = read_and_filter_excel(f, specific_column, date_column)
                    
            df_final[date_column] = pd.to_datetime(df_final[date_column])
            sorted_df = df_final.sort_values(by=date_column)
            
            combined_data = pd.concat([combined_data, sorted_df], ignore_index=True)   
            print("combined_data = ",combined_data)   
#     zf = zf_open.extractall()
    
    return combined_data

```

```python
#  EXTRA SHTUFF I DONT NEED

# for file_zip in files_zip:
#     zf = read_zip(file_zip)
#     print(zf)
#     print(f"Processing file: {file_zip}")
    
#     for file in zf:
#         print(file)
#         extracted_data = read_and_filter_excel(file, specific_column, date_column)
#         print(f"Processing file: {file}")
        
#         combined_data = pd.concat([combined_data, extracted_data], ignore_index=True)
        
        
#             with zipfile.ZipFile(zip_file_path, 'r') as z:
#         for filename in z.namelist():
#             # Assuming the file is a CSV, read it directly into a DataFrame
#             with z.open(filename) as f:
#                 df = pd.read_csv(f, usecols=[specific_column, date_column])
#                 df[date_column] = pd.to_datetime(df[date_column])
#                 sorted_df = df.sort_values(by=date_column)
#                 combined_data = pd.concat([combined_data, sorted_df], ignore_index=True)
#     return combined_data
# print(combined_data)
# print(len(combined_data))

# Iterate over the list of files and process each one
# for file in files:
#     print(f"Processing file: {file}")
#     extracted_data = read_and_filter_excel(file, specific_column, date_column)
    
#     combined_data = pd.concat([combined_data, extracted_data], ignore_index=True)
#     print(len(combined_data))


def read_and_filter_csv_from_zip(zip_file_path, specific_column, date_column):
    combined_data = pd.DataFrame()
    with zipfile.ZipFile(zip_file_path, 'r') as z:
        for filename in z.namelist():
            # Assuming the file is a CSV, read it directly into a DataFrame
            with z.open(filename) as f:
                df = pd.read_csv(f, usecols=[specific_column, date_column])
                df[date_column] = pd.to_datetime(df[date_column])
                sorted_df = df.sort_values(by=date_column)
                combined_data = pd.concat([combined_data, sorted_df], ignore_index=True)
    return combined_data

def process_standalone_csv(files, specific_column, date_column):
    combined_data = pd.DataFrame()
    for file_path in files:
        df = pd.read_csv(file_path, usecols=[specific_column, date_column])
        df[date_column] = pd.to_datetime(df[date_column])
        sorted_df = df.sort_values(by=date_column)
        combined_data = pd.concat([combined_data, sorted_df], ignore_index=True)
    return combined_data

# Correct the pattern to match ZIP files
pattern_zip = "C:\\Users\\ciand\\OneDrive\\Documents\\UCD Engineering\\Final Year Project\\Penmanshiel 2\\Pen*.zip"
files_zip = glob.glob(pattern_zip)

# Adjust `specific_column` and `date_column` to your actual column names
specific_column = 'Power (kW)'
date_column = '# Date and time'

combined_data_from_zip = pd.DataFrame()
for file_zip in files_zip:
    combined_data_from_zip = pd.concat([combined_data_from_zip, read_and_filter_csv_from_zip(file_zip, specific_column, date_column)], ignore_index=True)

# Continue with your pattern for standalone CSV files
pattern_csv = "C:\\Users\\ciand\\OneDrive\\Documents\\UCD Engineering\\Final Year Project\\Penmanshiel 2\\*\\Turbine_Data*.csv"
files_csv = glob.glob(pattern_csv)

combined_data_from_csv = process_standalone_csv(files_csv, specific_column, date_column)

# Combine data from both ZIP and CSV files
combined_data = pd.concat([combined_data_from_zip, combined_data_from_csv], ignore_index=True)
combined_data



data_filled = data_missing
data_filled['Power (kW)'] = output_final[0]
# data_filled.iloc[0:500]
print("data_filled = ", data_filled)

# data_filled_slice = data_filled[0:50000]
# print("data_filled_slice = ",data_filled_slice)

# data_filled_nulls = data_filled[data_filled['Power (kW)'].isnull()]
# data_filled_nulls_slice = data_filled_nulls.iloc[0:50000]
# print("data_filled_nulls_slice = ", data_filled_nulls_slice)

data_filled_null_count = data_filled.isnull().sum().sum()  
print('Number of null values:', data_filled_null_count)
```
